## Q5) Which team has choosen BOWL FIRST and won maximum times ?


```python
tosswinnbowlfirstgamewin=df[df['decision']=="BOWL FIRST"][df['toss_won']==df['winner']][['toss_won','decision','winner']]
```

```python
tosswinnbowlfirstgamewin
```

```python
tosswinnbowlfirstgamewin['winner'].value_counts()
```

```python
tosswinnbowlfirstgamewin['winner'].value_counts().plot.pie(figsize=(7,7),fontsize=16,autopct='%1.1f%%')
```

KKR is the team which has won maximum no of game after winning the toss the choosing the bowl first followed by MI. That means KKR is best at chasing the target in case of winning the toss.

## Similarly 

```python
df[df['decision']=="BAT FIRST"][df['toss_won']==df['winner']]['winner'].value_counts()
```

Where CSK is the team which has won maximum no of game after winning the toss the choosing the bat first followed by MI.

### BUT BUT BUT 
#### Lets see the team which have not won the toss but win the match most time while defending and chasing .

```python
# toss winner and winner are not same but toss winner(opponent)
# has choosen to bat first but still won the game.
df[df['decision']=="BAT FIRST"][df['toss_won']!=df['winner']]['winner'].value_counts()
```

CSK has won most no of games while chasing the target after losing the toss,followed by MI

### And

```python
# toss winner and winner are not same but toss winner(opponent)
# has choosen to bowl first but still won the game.
df[df['decision']=="BOWL FIRST"][df['toss_won']!=df['winner']]['winner'].value_counts()
```

```python
df[df['decision']=="BOWL FIRST"][df['toss_won']!=df['winner']]['winner'].value_counts().plot.pie(figsize=(9,9),fontsize=12,autopct='%1.1f%%')
plt.show()
```

MI has won most no of games while defending the target after losing the toss,followed by KXIP
